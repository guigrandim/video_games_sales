```python
# Performance Plataform VIsion
#P: O CEO precisa entender quais ecossistemas de hardware (PS5, Xbox Series, Switch, PC, etc.) são os mais lucrativos e diversificados
# identificar em qual plataforma vale mais a pena investir em exclusividades ou marketing

plataform_consoles = df1.loc[df1['console'] != 'PC']

#1.Qual o nome da plataforma que possui mais títulos únicos registrados?

count_plataform_unique_title = (
    df1.groupby('console')['title']
    .nunique()
)
count_console_unique_title = plataform_consoles.groupby('console')['title'].nunique()

plataform_most_unique = count_plataform_unique_title.idxmax()
console_most_unique = count_console_unique_title.idxmax()

print(f"A plataforma com mais titulos unicos é: {plataform_most_unique} com o total de: {count_plataform_unique_title.max()} titulos")
print(f"O console com mais titulos únicos é: {console_most_unique} com o total de: {count_console_unique_title.max()} titulos" )

#2.Qual o nome da plataforma que possui mais jogos com Critic Score médio acima de 80?

mean_plataform_titles = (
    df1.groupby(['console', 'title'])['critic_score']
    .mean()
    .reset_index(name='mean_critic_score')
)

plataform_titles_80 = mean_plataform_titles[mean_plataform_titles['mean_critic_score'] >= 8.0]

mean_console_titles = plataform_consoles.groupby(['console', 'title'])['critic_score'].mean().reset_index(name='mean_critic_score')
console_titles_80 = mean_console_titles[mean_console_titles['mean_critic_score'] >= 8.0]

count_titles_plataform_80 = plataform_titles_80.groupby('console')['title'].count().sort_values(ascending = False)
count_titles_console_80 = console_titles_80.groupby('console')['title'].count().sort_values(ascending = False)

plataform_best = count_titles_plataform_80.idxmax()
console_best = count_titles_console_80.idxmax()

print(f"A plataforma com mais titulos acima de 8.0 no critic score é: {plataform_best} com {count_titles_plataform_80.max()}")
print(f"O console com mais titulos acima de 8.0 no critic score é: {console_best} com {count_titles_console_80.max()}")

#3.Qual o nome da plataforma que possui mais jogos com Critic Score médio abaixo de 40?

plataform_titles_40 = mean_plataform_titles[mean_plataform_titles['mean_critic_score'] <= 4.0]
count_titles_plataform_40 = (
    plataform_titles_40.groupby('console')['title']
    .count()
    .sort_values(ascending = False)
)

plataform_worst = count_titles_plataform_40.idxmax()

print(f"A plataforma com mais titulos abaixo de 4.0 no critic score é: {plataform_worst} com {count_titles_plataform_40.max()}")

#4.Qual o nome da plataforma que possui o maior valor médio de vendas globais por título?

plataform_title_sales_global = (
    df1.groupby('console')['total_sales']
    .mean()
    .sort_values(ascending = False)
    .reset_index(name='mean_sales_per_title')
    .round(2)
)

plataform_title_sales_global_best = plataform_title_sales_global.iloc[0]['console']
plataform_sales_global_best = plataform_title_sales_global.iloc[0]['mean_sales_per_title']

print(plataform_title_sales_global)

print(f"O maior valor médio de vendas globais por titulo é {plataform_title_sales_global_best} com {plataform_sales_global_best}")

#5.Qual o nome da plataforma que possui a maior quantidade de gêneros distintos em seu catálogo?

plataform_genre_count = df1.groupby('console')['genre'].nunique().sort_values(ascending = False)
max_genre = plataform_genre_count.max()

plataform_genre_count[plataform_genre_count == max_genre]

#6. Qual o nome da plataforma que possui o maior volume acumulado de vendas na América do Norte (NA)?

plataform_sum_sales_na = (
    df1.groupby('console')['na_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name='sum_sales_per_console')
    .round(2)
)

best_console_na = plataform_sum_sales_na.iloc[0]['console']
total_sales_console_na = plataform_sum_sales_na.iloc[0]['sum_sales_per_console']

print(f'O console com mais vendas na America do Norte foi: {best_console_na} com total de {total_sales_console_na} milhões de unidades')

#7. Qual o nome da plataforma que possui o maior volume acumulado de vendas no Japão (JP)?

plataform_sum_sales_jp = (
    df.groupby('console')['jp_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name='sum_sales_per_console')
    .round(2)
)

best_console_jp = plataform_sum_sales_jp.iloc[0]['console']
total_sales_console_jp = plataform_sum_sales_jp.iloc[0]['sum_sales_per_console']

print(f'O console com mais vendas no Japão foi: {best_console_jp} com o total de {total_sales_console_jp} milhões de unidades')

#8. Qual o nome da plataforma que possui o maior volume acumulado de vendas no Europa/África (PAL)?

plataform_sum_sales_pal = (
    df1.groupby('console')['pal_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name='sum_sales_per_console')
    .round(2)
)

best_console_pal = plataform_sum_sales_pal.iloc[0]['console']
total_sales_console_pal = plataform_sum_sales_jp.iloc[0]['sum_sales_per_console']

print(f'O console com mais vendas na Europa/África foi: {best_console_pal} com o total de {total_sales_console_pal} milhoes de unidades')

#9. Qual plataforma mais vendida por geração ?

plataform_sum_sales_global_gen = (
    plataform_consoles.groupby(['generation','console', 'manufacture'])['total_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name='sales_per_generation')
)

all_gen_sales = plataform_sum_sales_global_gen.groupby('generation').first().reset_index()
all_gen_sales = all_gen_sales[(all_gen_sales['generation'] != '9th Gen') & (all_gen_sales['generation'] != 'Other/Unknown')]
print(all_gen_sales)

#10. Qual plataforma teve mais developer e publishers únicos desenvolvendo titulos ?

developers_count_per_plataform = df1.groupby('console')['developer'].nunique().sort_values(ascending = False)
developers_count_per_console = plataform_consoles.groupby('console')['developer'].nunique().sort_values(ascending = False)

publishers_count_per_plataform = df1.groupby('console')['publisher'].nunique().sort_values(ascending = False)
publishers_count_per_console = plataform_consoles.groupby('console')['publisher'].nunique().sort_values(ascending = False)

best_plataform_developers = developers_count_per_plataform.idxmax()
best_console_developers = developers_count_per_console.idxmax()

best_plataform_publishers = publishers_count_per_plataform.idxmax()
best_console_publishers = publishers_count_per_console.idxmax()

print(f'A principal plataforma para desenvolvedores é: {best_plataform_developers} com {developers_count_per_plataform.max()} e para distribuidores é {best_plataform_publishers} com {publishers_count_per_plataform.max()}')
print(f'A principal plataforma para consoles é: {best_console_developers} com {developers_count_per_console.max()} e para distribuidores é {best_console_publishers} com {publishers_count_per_console.max()}')
```
