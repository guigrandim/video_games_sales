```python
# Performance Plataform VIsion
#P: O CEO precisa identificar os "cavalos vencedores" do catálogo
# entender se a qualidade técnica (notas) garante retorno financeiro e como gêneros específicos se comportam em mercados distintos.
# O CEO quer saber onde estão as falhas de desenvolvimento (notas baixas) e os fenômenos de vendas.

#1. Qual o nome do jogo que possui a maior quantidade de Vendas Globais?

sales_global_per_title = (
    df1.groupby('title')['total_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name = 'total_global_sales')
)

best_title_sales = sales_global_per_title.iloc[0]['title']
qtd_title_sales = sales_global_per_title.iloc[0]['total_global_sales']

print(f'O jogo que possui a maior quantidade de vendas globais é {best_title_sales} com {qtd_title_sales} milhões de unidades')

#2. Qual o nome do jogo com o maior Critic Score registrado?

max_critic_score = df1['critic_score'].max()
best_critc_score_ever = df1[df1['critic_score'] == max_critic_score][['title','critic_score','developer']]
title_game = best_critc_score_ever.iloc[0]['title']
score_game = best_critc_score_ever.iloc[0]['critic_score']
developer_game = best_critc_score_ever.iloc[0]['developer']

print(f'O jogo com o maior Critic Score registrado é {title_game} com nota {score_game} da desenvolvedora {developer_game}')

#3. Qual o nome do jogo que possui o maior volume de vendas apenas na América do Norte (NA)?

sales_title_na = (
    df1.groupby('title')['total_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name = 'total_sales_na')
)

best_title_sales = sales_title_na.iloc[0]['title']
qtd_title_sales = sales_title_na.iloc[0]['total_sales_na']

print(f'O jogo que possui o maior volume de vendas apenas na América do Norte é {best_title_sales} com {qtd_title_sales} milhoes de unidades')

#4.Qual o nome do jogo do gênero "Sports" que possui o menor Critic Score?

sport_critic_score = df1[df1['genre'] == 'Sports'].sort_values('critic_score', ascending = True)
name_title_min_score = sport_critic_score['title'].iloc[0]
sport_title_critic_score = sport_critic_score['critic_score'].iloc[0]
print(f'O jogo do gênero "Sports" que possui o menor Critic Score {name_title_min_score} com nota igual a {sport_title_critic_score}')

#5.Qual o nome do jogo do gênero "Action", lançado para a plataforma PS4, que possui o maior Critic Score?

action_ps5_critic_score = df1[(df1['genre'] == 'Action') & (df1['console'] == 'XOne')].sort_values('critic_score', ascending = False)
name_title_max_score = action_ps5_critic_score['title'].iloc[0]
action_title_critic_score_max = action_ps5_critic_score['critic_score'].iloc[0]

print(f'O jogo do gênero "Action" lançado para a plataforma XBOX One que possui o maior Critic Score: {name_title_max_score} com nota igual a {action_title_critic_score_max}')

#6. Jogos lançados para múltiplas plataformas possuem, na mediana, um volume de vendas maior que jogos exclusivos (lançados para apenas uma plataforma)?

plataforms_per_title = df1[df1['console'] != 'All'].groupby('title')['console'].nunique()
df1['type'] = df1['title'].map(plataforms_per_title).apply(lambda x: 'multiplataforma' if x > 1 else 'exclusivo')

median_sales_per_type = (
    df1.groupby('type')['total_sales']
    .agg(median_sales = ('median'))
    .reset_index()
)
print(median_sales_per_type)

#7.Jogos com Critic Score acima de 90 são também, na média, os títulos que possuem o maior volume de vendas globais?

premium_critic_score = df1[df1['classification'] == 'Premium']['total_sales'].mean().round(2)
rest_critic_score = df1[df1['classification'] != 'Premium']['total_sales'].mean().round(2)

print(f'Jogos com Critic Score acima de 90 tem {premium_critic_score} milhoes de vendas enquanto abaixo apresenta {rest_critic_score} milhoes de vendas')

#8.Os jogos do gênero "Role-Playing" (RPG) lançados no Japão (JP) possuem um valor médio de vendas maior que os jogos de "Shooter" da América do Norte (NA)?

compare_sales_per_genre_and_region = pd.Series({
    'jp_rpg' : df1[df1['genre'] == 'Role-Playing']['jp_sales'].mean(),
    'na_shooter' : df1[df1['genre'] == 'Shooter']['na_sales'].mean()
}).sort_values(ascending = False).reset_index(name = 'mean_sales').round(2)

mean_sales_jp = compare_sales_per_genre_and_region['mean_sales'].iloc[1]
mean_sales_na = compare_sales_per_genre_and_region['mean_sales'].iloc[0]

print(f'Jogos do genero RPG lancados no JP possuem um valor médio de vendas {mean_sales_jp} mil cópias enquanto o Shooter nos NA {mean_sales_na} mil cópias')

#9. Quais os titulos mais vendidos globalmente por geração (até a 8th) ?

clean_gen = df1[(df1['generation'] != '9th Gen') & (df1['generation'] != 'Other/Unknown')]

sales_title_per_gen = (
    clean_gen.groupby(['generation','title'])['total_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name = 'sales_per_gen')
)

sales_top_title_per_gen = sales_title_per_gen.groupby('generation').first().reset_index()
sales_top_title_per_gen

sales_title_per_publisher = (
    df1.groupby(['publisher','title', 'release_date'])['total_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name = 'sales_per_publisher')
)

clean_publishers = sales_title_per_publisher[sales_title_per_publisher['sales_per_publisher'] > 0]

sales_top_title_per_publisher = clean_publishers.groupby('publisher').first().reset_index()
sales_top_title_per_publisher
```
