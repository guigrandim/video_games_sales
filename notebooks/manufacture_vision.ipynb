```python
# Performance Manufacture Vision
#P: O CEO precisa entender o Market Share por Marca e o Ciclo de Vida das Gerações
#O: O objetivo é identificar qual fabricante detém a maior lealdade de público e qual geração foi o "pico de ouro" da indústria.

#Create life cycles manufacture
start_life_cycle_manufacture = df1.groupby('manufacture')['start_year'].min().reset_index(name = 'start_year')
end_life_cycle_manufacture = df1.groupby('manufacture')['end_year'].max().reset_index(name = 'end_year')

life_cycle_manufacture = start_life_cycle_manufacture.merge(end_life_cycle_manufacture, on='manufacture', how = 'inner')

life_cycle_manufacture['activity_years'] = life_cycle_manufacture['end_year'] - life_cycle_manufacture['start_year']
life_cycle_manufacture

activity_map = life_cycle_manufacture.set_index('manufacture')['activity_years']

#Filter Global Sales
filter_sales_global = df1[df1['total_sales'] > 0.1]

#1. Qual o nome do Fabricante (Manufacturer) que possui o maior volume de vendas globais acumulado?

manufacture_sales_global_history = (
    filter_sales_global.groupby(['manufacture'])['total_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name = 'total_sales_history')
)

manufacture_sales_global_history = manufacture_sales_global_history[manufacture_sales_global_history['manufacture'] != 'Other/Unknown']

manufacture_sales_global_history['activity_years'] = manufacture_sales_global_history['manufacture'].map(activity_map)

manufacture_sales_global_history['mean_per_activity_years'] = (
    manufacture_sales_global_history['total_sales_history'] / manufacture_sales_global_history['activity_years']
).round(2)

best_manufacture_history = manufacture_sales_global_history['manufacture'].iloc[0]
best_manufacture_sales = manufacture_sales_global_history['mean_per_activity_years'].iloc[0]

print(f'O fabricante que possui o maior volume de vendas globais por tempo de atividade'
      f'é: {best_manufacture_history} com {best_manufacture_sales} milhoes de vendas')

#2. Qual fabricante possui a maior quantidade de títulos com Critic Score médio acima de 85?

manufacture_critic_85 = filter_sales_global[filter_sales_global['critic_score'] >= 8.5]

manufacture_sales_per_critic = (
    manufacture_critic_85.groupby('manufacture')['total_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name = 'total_sales_critic')
)

best_manufacture_per_critic = manufacture_sales_per_critic['manufacture'].iloc[0]
best_manufacture_sales_critic = manufacture_sales_per_critic['total_sales_critic'].iloc[0]

print(f'A fabricante possui a maior quantidade de títulos com Critic Score médio acima de 85 é: {best_manufacture_per_critic} com {best_manufacture_sales_critic} milhões de venda')

#3. Qual fabricante detém a maior fatia de mercado (Vendas) no Japão (JP) em comparação com a América do Norte (NA)?

clean_unknown_manufacture = df1[df1['manufacture'] != 'Other/Unknown']

na_sales_manufactures = (
    clean_unknown_manufacture.groupby('manufacture')['na_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name = 'total_sales_na')  
)

na_sales_manufactures_filter = na_sales_manufactures[na_sales_manufactures['total_sales_na'] > 0.1]

jp_sales_manufactures = (
    clean_unknown_manufacture.groupby('manufacture')['jp_sales']
    .sum()
    .sort_values(ascending = False)
    .reset_index(name = 'total_sales_jp')
)

jp_sales_manufactures_filter = jp_sales_manufactures[jp_sales_manufactures['total_sales_jp'] > 0.1]
jp_na_sales = na_sales_manufactures_filter.merge(jp_sales_manufactures_filter, on='manufacture', how='inner')
jp_na_sales['activity_years'] = jp_na_sales['manufacture'].map(activity_map)

jp_na_sales['sales_per_activity_na'] = jp_na_sales['total_sales_na'] / jp_na_sales['activity_years']
jp_na_sales['sales_per_activity_jp'] = jp_na_sales['total_sales_jp'] / jp_na_sales['activity_years']

jp_na_sales['diference_na_jp'] = (jp_na_sales['sales_per_activity_na'] / jp_na_sales['sales_per_activity_jp'])
jp_manufacture_domain = jp_na_sales['manufacture'].iloc[-1]

print(f'A fabricante que detém a maior fatia de mercado (Vendas) no Japão (JP) em '
      f'comparação com a América do Norte (NA) é a {jp_manufacture_domain}')

#4. Qual fabricante lançou a maior diversidade de consoles (quantidade de console_name diferentes) ao longo da história?

clean_unknown_manufacture = df1[df1['manufacture'] != 'Other/Unknown']

release_consoles_manufacture = (
    clean_unknown_manufacture.groupby('manufacture')
    .agg(
        total_release = ('console', 'nunique'),
        total_sales = ('total_sales', 'sum')
    )
    .reset_index()
)

release_consoles_manufacture['activity_years'] = release_consoles_manufacture['manufacture'].map(activity_map)

release_consoles_manufacture['sales_per_activity_years'] = (
    release_consoles_manufacture['total_sales'] / release_consoles_manufacture['activity_years']
).round(2)

release_consoles_manufacture['efficiency_per_console'] = (
    release_consoles_manufacture['total_release'] / release_consoles_manufacture['sales_per_activity_years']
).round(2)

manufacture_best_console_release = release_consoles_manufacture.loc[release_consoles_manufacture['total_release'].idxmax(), 'manufacture']
manufacture_qtd_console_release = release_consoles_manufacture.loc[release_consoles_manufacture['total_release'].idxmax(), 'total_release']
manufacture_efficience_release = release_consoles_manufacture.loc[release_consoles_manufacture['efficiency_per_console'].idxmin(), 'manufacture']

print(f'O fabricante que lançou a maior diversidade de consoles da historia foi {manufacture_best_console_release} com {manufacture_qtd_console_release} consoles '
      f'enquanto o fabricante mais eficiente foi {manufacture_efficience_release}')

#5. Qual o fabricante que possui a maior média de vendas por jogo lançado?

clean_release_date = df1.dropna(subset=['release_date'])
clean_release_date = clean_release_date[clean_release_date['manufacture'] != 'Other/Unknown']

total_sales_per_games_release = (
    clean_release_date.groupby('manufacture')
    .agg(
        release_games = ('release_date', 'count'),
        total_sales = ('total_sales', 'sum')
    )
    .reset_index()
)

total_sales_per_games_release['sales_per_game_release'] = (total_sales_per_games_release['total_sales'] / total_sales_per_games_release['release_games']).round(2)
total_sales_per_games_release = total_sales_per_games_release[total_sales_per_games_release['sales_per_game_release'] > 0]

best_manufacture_sales_release_game = total_sales_per_games_release.loc[total_sales_per_games_release['sales_per_game_release'].idxmax(), 'manufacture']

print(f'O fabricante que possui a maior média de vendas por jogo lançado é {best_manufacture_sales_release_game}')

#6. Qual Geração (Generation) de consoles foi a mais lucrativa em termos de vendas globais?

filter_sales_global_gen = df1[(df1['total_sales'] > 0.1) & (df1['generation'] != 'Other/Unknown')]

sales_global_per_generation = (
    filter_sales_global_gen.groupby('generation')['total_sales']
    .sum()
    .reset_index()
)

best_generation_sales = sales_global_per_generation.loc[sales_global_per_generation['total_sales'].idxmax(), 'generation']
total_generation_sales = sales_global_per_generation.loc[sales_global_per_generation['total_sales'].idxmax(), 'total_sales']

print(f'A geração de consoles foi a mais lucrativa em termos de vendas globais foi: {best_generation_sales} com {total_generation_sales}'
      f' milhoes de unidades vendidas')

#7. Existe uma tendência de melhora nos Critic Scores médios à medida que as gerações avançam (comparando a 7ª, 8ª e 9ª gerações)?

filter_critic_score_gen = df1[(df1['critic_score'] > 0) & (df1['generation'] != 'Other/Unknown')]

critic_score_per_generation = (
    filter_critic_score_gen.groupby('generation')
    .agg(
        critic_score_mean = ('critic_score', 'mean'),
        critic_score_median = ('critic_score', 'median'),
        critic_score_std = ('critic_score', 'std'),
        critic_score_count = ('critic_score', 'count')
    )
    .reset_index()
)

critic_score_per_generation = critic_score_per_generation[critic_score_per_generation['critic_score_count'] > 50]
critic_score_per_generation['cv_critic_score'] = ((
    critic_score_per_generation['critic_score_std'] / critic_score_per_generation['critic_score_mean']
    ) * 100 ).round(0)

critic_score_per_generation['pct_variation'] = ((
    (critic_score_per_generation['critic_score_mean'] - critic_score_per_generation['critic_score_mean'].shift(1)) / critic_score_per_generation['critic_score_mean'].shift(1)
) * 100 ).round(1)

critic_score_7th_gen_pct = critic_score_per_generation['pct_variation'].iloc[3]
critic_score_8th_gen_pct = critic_score_per_generation['pct_variation'].iloc[4]

print(f'Existe uma tendencia de melhora a partir da 7th Gen com um aumento de {critic_score_7th_gen_pct}% percentual nas avaliações ' 
      f'consolidando na 8th geração com {critic_score_8th_gen_pct}% de melhora nas avaliações')

#8. Qual geração possui a maior quantidade de gêneros diferentes de jogos registrados?

filter_gen = df1[df1['generation'] != 'Other/Unknown']

gen_genere_game_unique = filter_gen.groupby('generation')['genre'].nunique().sort_values(ascending = False).reset_index()

best_gen = gen_genere_game_unique[gen_genere_game_unique['genre'] == gen_genere_game_unique['genre'].max()]

print(f'As gerações que possuem a maior quantidade de gêneros diferentes de jogos registrados são:\n {best_gen}')

#9. Qual geração de hardware possui a maior concentração de jogos do gênero "RPG"?

filter_gen_genre = df1[(df1['generation'] != 'Other/Unknown') & (df1['genre'] == 'Role-Playing')]

rpg_count_gen = filter_gen_genre.groupby('generation')['title'].nunique().reset_index(name = 'total_titles')

best_gen_rpg = rpg_count_gen.loc[rpg_count_gen['total_titles'].idxmax(), 'generation']
best_gen_rpg_count = rpg_count_gen.loc[rpg_count_gen['total_titles'].idxmax(), 'total_titles']

print(f'A geração de hardware que possui a maior concentração de jogos do gênero "RPG" é {best_gen_rpg} com {best_gen_rpg_count}')

#10. Entre todos os console, qual família de hardware (ex: todos os "PlayStation") possui a maior nota média dos usuários?
#11. Qual console_name específico foi o mais bem-sucedido de um fabricante que não seja do "Big Three" (Sony, Nintendo, Microsoft)?

filter_gen_genre = df1[(df1['manufacture'] != 'Other/Unknown') & (df1['critic_score'] > 0)]

critic_score_manufactrue_per_class = (
    filter_gen_genre.groupby(['manufacture', 'console'])['critic_score']
    .mean()
    .reset_index()
)

best_critic_score_manufactrues = (
    critic_score_manufactrue_per_class.loc
    [critic_score_manufactrue_per_class.groupby('manufacture')['critic_score'].idxmax()]
    [['manufacture','console','critic_score']]
).round(1).reset_index(drop=True)

print(best_critic_score_manufactrues)

big_three_drop = ['Nintendo', 'Sony', 'Microsoft', 'Apple']

best_critic_score_manufactrue_wb3 = best_critic_score_manufactrues[~best_critic_score_manufactrues['manufacture'].isin(big_three_drop)].sort_values('critic_score', ascending = False)

print(f"O console mais bem-sucedido de um fabricante que não seja do Big Three é {best_critic_score_manufactrue_wb3['console'].values[0]}")
```
